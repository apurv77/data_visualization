```python
import matplotlib.pyplot as plt
%matplotlib inline
#show plots
import numpy as np
import pandas as pd

%reload_ext autoreload
%autoreload 2
#reload page

#chart 1

x_mb_s3 = ([0.1, 1, 10])
y_time_s3 = ([1.35, 2.23, 5.14])

#x_mb_el = ([10, 50, 100])
#y_time_el = ([7, 14, 21])

x_mb_k8s = ([0.1, 1, 10])
y_time_k8s = ([0.45, 0.5, 2.6])
```

```python
size_vs_extime = plt.figure(figsize=(7,6), dpi=100)
sz_ex_axes = size_vs_extime.add_axes([0.1, 0.1, 0.9, 0.9])
sz_ex_axes.set_xlabel('Size(mb)')
sz_ex_axes.set_ylabel('Execution Time(s)')
sz_ex_axes.set_title('Size vs Execution Time')
sz_ex_axes.plot(x_mb_s3, y_time_s3, label='Lambda S3', color='red', alpha=.75, lw=2, ls='-.',
           marker='s', markersize=7, markerfacecolor='black',
           markeredgecolor='black', markeredgewidth=2)
#sz_ex_axes.plot(x_mb_el, y_time_el, label='Lambda ElastiCache', color='orange', alpha=.75, lw=2, ls='-.',
#           marker='s', markersize=7, markerfacecolor='black',
#           markeredgecolor='black', markeredgewidth=2)
sz_ex_axes.plot(x_mb_k8s, y_time_k8s, label='k8s', color='blue', alpha=.75, lw=2, ls='-.',
           marker='s', markersize=7, markerfacecolor='black',
           markeredgecolor='black', markeredgewidth=2)
sz_ex_axes.legend(loc=0)
size_vs_extime.savefig('SzVsEx.png')
```

```python
#chart 2 (size vs memory usage)
x_mb_s3_2 = ([0.1, 1, 10])
y_time_s3_2 = ([18, 22, 54])

x_mb_k8s_2 = ([0.1, 1, 10])
y_time_k8s_2 = ([47, 53, 80])
```
